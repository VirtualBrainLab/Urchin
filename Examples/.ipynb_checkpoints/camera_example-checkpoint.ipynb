```python
urn.setup()
```

```python
urn.clear()

urn.set_area_visibility({'grey':True})
urn.set_area_material({'grey':'transparent-unlit'})
urn.set_area_alpha({'grey':0.5})

areas = ['MRN','SCm', 'SCs', 'APN']
for area in areas:
    urn.set_area_visibility({area:True})
    urn.set_area_material({area:'default'})
```
